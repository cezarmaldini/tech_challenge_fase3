### **IMPORTAR BIBLIOTECAS**

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

### **LER DADOS DA CAMADA SILVER**

```python
df = spark.read \
          .format('parquet') \
          .load('Files/silver/dados')
```

### **ESTADOS E CAPITAIS**

```python
df_uf = spark.read \
             .format('parquet') \
             .load('Files/silver/dicionarios/dict_uf')

df.createOrReplaceTempView('df_tmp1')
df_uf.createOrReplaceTempView('uf_tmp')

query = """
SELECT
    d.*,
    u.estado, u.capital
FROM
    df_tmp1 d
LEFT JOIN
    uf_tmp u
ON
    d.id = u.id    
"""

df = spark.sql(query)
```

### **CARACTERÍSTICAS GERAIS DOS MORADORES**

```python
df_caract = spark.read \
                 .format('parquet') \
                 .load('Files/silver/dicionarios/dict_caracteristicas')
```

#### **SEXO**

```python
df_sexo = df_caract.filter(col('pergunta') == 'Sexo') \
                   .select('respostaId', 'resposta') \
                   .distinct()

df.createOrReplaceTempView('df_tmp2')
df_sexo.createOrReplaceTempView('sexo_tmp')

query = """
SELECT
    d.*,
    s.resposta AS sexo
FROM
    df_tmp2 d
LEFT JOIN
    sexo_tmp s
ON
    d.sexoId = s.respostaId   
"""

df = spark.sql(query)
```

#### **COR OU RAÇA**

```python
df_cor = df_caract.filter(col('pergunta') == 'Cor ou raça') \
                  .select('respostaId', 'resposta') \
                  .distinct()

df.createOrReplaceTempView('df_tmp3')
df_cor.createOrReplaceTempView('cor_tmp')

query = """
SELECT
    d.*,
    c.resposta AS cor
FROM
    df_tmp3 d
LEFT JOIN
    cor_tmp c
ON
    d.corId = c.respostaId   
"""

df = spark.sql(query)
```

#### **ESCOLARIDADE**

```python
df_escolaridade = df_caract.filter(col('pergunta') == 'Escolaridade') \
                           .select('respostaId', 'resposta') \
                           .distinct()

df.createOrReplaceTempView('df_tmp4')
df_escolaridade.createOrReplaceTempView('escolaridade_tmp')

query = """
SELECT
    d.*,
    e.resposta AS escolaridade
FROM
    df_tmp4 d
LEFT JOIN
    escolaridade_tmp e
ON
    d.escolaridadeId = e.respostaId   
"""

df = spark.sql(query)
```

### **CARACTERÍSTICAS DA HABITAÇÃO**

```python
df_habitacao = spark.read \
                    .format('parquet') \
                    .load('Files/silver/dicionarios/dict_habitacao')
```

#### **DOMICÍLIO**

```python
df_domicilio = df_habitacao.filter(col('id') == 'F001') \
                           .select('respostaId', 'resposta') \
                           .distinct()

df_domicilio.createOrReplaceTempView('domicilio_tmp')
df.createOrReplaceTempView('df_tmp12')

query = """
SELECT
    d.*,
    do.resposta AS domicilio
FROM
    df_tmp12 d
LEFT JOIN
    domicilio_tmp do
ON
    d.domicilioId = do.respostaId   
"""

df = spark.sql(query)
```

#### **VALOR ALUGUEL**

```python
df_aluguel = df_habitacao.filter(col('id') == 'F0022') \
                         .select('respostaId', 'resposta') \
                         .distinct()

df_aluguel.createOrReplaceTempView('aluguel_tmp')
df.createOrReplaceTempView('df_tmp13')

query = """
SELECT
    d.*,
    a.resposta AS valor_aluguel
FROM
    df_tmp13 d
LEFT JOIN
    aluguel_tmp a
ON
    d.valor_aluguelId = a.respostaId   
"""

df = spark.sql(query)
```

### **TRANSFORMAÇÕES FINAIS**

```python
df = df.select(
    'ano', 'mes', 'estado', 'capital', 'idade', 'sexo', 'cor', 'escolaridade', 'exame_swab', 'exame_dedo',
    'exame_sangue', 'trabalho', 'faixa_rendimentoId', 'faixa_renda', 'bolsa_familia', 'auxilio', 'domicilio', 'valor_aluguel'
    ) \
    .withColumn("data",
        to_date(
            concat_ws("-", col("ano"), lpad(col("mes").cast("string"), 2, "0"), lit("01")),
            "yyyy-MM-dd"
        )
    ) \
    .withColumn(
        'mes_nome',
        when(col('mes') == 9, 'setembro')
        .when(col('mes') == 10, 'outubro')
        .when(col('mes') == 11, 'novembro')
    ).distinct()

display(df)
```

### **ARMAZENAR DADOS**

```python
df.write \
  .format('delta') \
  .mode('overwrite') \
  .saveAsTable('dados')

print('Dados armazenados com sucesso na camada gold! 🥇')
```
